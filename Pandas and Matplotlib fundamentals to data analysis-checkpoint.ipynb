# Importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Opening the file in Pandas

```python
df_food_waste = pd.read_csv("food_waste.csv")
df_food_waste
```

## Selecting the columns to make analysis

```python
df_food_waste = df_food_waste[['Year', 'Entity', '12.3.1 - Food waste per capita (KG) - AG_FOOD_WST_PC - Households']]
```

## Rename the columns to easily understand

```python
df_food_waste = df_food_waste.rename(columns={'Entity':'Country', 
                                '12.3.1 - Food waste per capita (KG) - AG_FOOD_WST_PC - Households':'Food_waste_Kg_percapita'})
df_food_waste
```

## Sorting by food waste values

```python
df_food_waste = df_food_waste.sort_values('Food_waste_Kg_percapita', ascending=False)
df_food_waste
```

## Using .pivot() method to select which country we want to analyse

```python
df_pivot = df_food_waste.pivot(index='Year', columns='Country', values="Food_waste_Kg_percapita")
df_pivot
```

## Selecting the countries

```python
df_pivot = df_pivot[['Nigeria', 'Rwanda', 'Greece', 'Bahrain', 'Malta', 'Australia', 'Brazil', 'United States', 
                     'China', 'Japan', 'India', 'Austria', 'Russia', 'World']]
```

## Making a barchart

```python
df_pivot = df_pivot.T
```

```python
df_pivot.plot(kind='bar', ylabel= 'KG_PerCapita', title='Food Waste 2019 - Households', figsize=(8,4))
#save the plot
plt.savefig('Food_waste_2019')
#.show() method
plt.show()
```
